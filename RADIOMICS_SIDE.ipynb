#EXTRACTING FEATURES FROM SIDES

```python
import os
import numpy as np
import radiomics
from radiomics import featureextractor
import SimpleITK as sitk
import cv2

# Set the path to the folder containing your images
image_folder = '/content/drive/MyDrive/concatenatedzip/validation/Erosion'
# Changed output folder to a folder within the user's drive
output_folder = '/content/drive/MyDrive/validation/output_Erosion_sides'

# Make the output folder if it doesn't already exist
!mkdir -p {output_folder}

# Initialize the radiomics feature extractor
# You can specify settings in the parameter file (YAML) or directly in the extractor
extractor = featureextractor.RadiomicsFeatureExtractor()

# Iterate over all images in the folder
for image_name in os.listdir(image_folder):
    if image_name.endswith('.jpg'):
        image_path = os.path.join(image_folder, image_name)

        # Read the image using OpenCV and convert it to grayscale
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        # Convert the image to a SimpleITK image (required by PyRadiomics)
        sitk_image = sitk.GetImageFromArray(image)

        # Create a mask with a distinct value for the segmented region
        mask = np.ones(image.shape, dtype=np.uint8)
        # Example: set the center region of the mask to 1
        center_x, center_y = image.shape[1] // 2, image.shape[0] // 2
        mask[center_y - 50:center_y + 50, center_x - 50:center_x + 50] = 0

        sitk_mask = sitk.GetImageFromArray(mask)

        # Extract features
        features = extractor.execute(sitk_image, sitk_mask)

        # Save the features to a file
        feature_file = os.path.join(output_folder, image_name.replace('.jpg', '.npy'))
        np.save(feature_file, features)

        print(f"Features extracted and saved for {image_name}")

print("Feature extraction complete.")
```

#COCATENATE FEATURES

```python
import numpy as np
import pandas as pd
import os

# Set the path to the folder containing your feature files
feature_folder = '/content/drive/MyDrive/validation/output_worm_sides'

# Initialize lists to store feature data and image names
features_list = []
image_names = []

# Load feature files
for feature_file in os.listdir(feature_folder):
    if feature_file.endswith('.npy'):
        feature_path = os.path.join(feature_folder, feature_file)
        features = np.load(feature_path, allow_pickle=True).item()  # Load the dictionary
        features_list.append(features)
        image_names.append(feature_file.replace('.npy', '.jpg'))  # Assuming you know the image names

# Convert features to DataFrame
df_features = pd.DataFrame(features_list)
df_features['image_name'] = image_names

# Save the DataFrame to a CSV file for later use
df_features.to_csv('/content/drive/MyDrive/output_validation/features_worm_sides.csv', index=False)

print("Features have been processed and saved to 'features.csv'.")
```
