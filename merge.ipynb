```python
import os
import pandas as pd

# Define the list of disease names
disease_names = ['Lymphangiectasia', 'Polyp', 'Angioectasia', 'Bleeding', 'Erosion',
                 'erythema', 'fb', 'ulcer', 'worm','Normal']

# Path to the folder where the CSV files are stored
output_folder = '/content/'

# List to store dataframes
dfs = []

# Iterate through each disease, load the corresponding combined CSV, and store in a list
for disease in disease_names:
    # Construct the filename for each disease
    file_path = os.path.join(output_folder, f'updated_features_{disease}.csv')

    if os.path.exists(file_path):
        # Read the CSV file for this disease
        df = pd.read_csv(file_path)

        # Append the dataframe to the list
        dfs.append(df)
    else:
        print(f"File {file_path} does not exist!")

# Ensure that all files have been loaded correctly
if dfs:
    # Horizontally concatenate all dataframes based on their index (row-wise concatenation)
    stacked_df = pd.concat(dfs, axis=0)

    # Save the stacked dataframe as a new CSV
    stacked_filename = 'horizontally_stacked_features.csv'
    stacked_df.to_csv(os.path.join(output_folder, stacked_filename), index=False)

    print(f"Horizontally stacked features saved as {stacked_filename}")
else:
    print("No valid files were found for stacking.")
```
